```python
# df = load_iris(return_X_y=True, as_frame=True)
df = load_wine(return_X_y=True, as_frame=True)
```

```python
df_all = df[0]
df_all['y'] = df[1].values
```

```python
df_all
```

```python

convert_dict = {'y': str}
 
df_all = df_all.astype(convert_dict)

df_all.dtypes
```

```python
df_all
```

```python
df_all.dtypes
```

```python
df_all.dtypes.values[0] == float
```

```python
mystr = "["
for index, row in df_all.iterrows():
    mystr = mystr + "["
    for col in range(df_all.shape[1]):
        if df_all.dtypes.values[col] == float:
            mystr = mystr + "#number(" + f'{row[col]:.1f}' + ")"
        else:
            mystr = mystr + "#symbol(" + '\"' + f'{int(row[col])}' + '\"' + ")"
        if col < df_all.shape[1]-1:
            mystr = mystr + ", "
    mystr = mystr + "]"

    if index < df_all.shape[0]-1:
        mystr = mystr + ",\n"
mystr = mystr + "]"
print(mystr)
```
