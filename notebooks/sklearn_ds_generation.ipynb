```python
# Regression tree example with diabetes dataset
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.tree import DecisionTreeRegressor
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
random.seed(1234567)
X, y = datasets.load_diabetes(return_X_y=True)
pos = random.sample(range(X.shape[0]), X.shape[0])

xtrain = X[pos[:300],:]
ytrain = y[pos[:300]] 
xtest = X[pos[300:],:]
ytest = y[pos[300:]] 

dtr = DecisionTreeRegressor(max_depth=10,min_samples_split=5)
dtr.fit(xtrain, ytrain)
yhat = dtr.predict(xtest)
print(np.mean(np.sqrt((yhat-ytest)**2)))

# Random Forest Regression example with diabetes dataset
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=100,max_depth=10,min_samples_split=5)
rfr.fit(xtrain, ytrain)
yhat = rfr.predict(xtest)
print(np.mean(np.sqrt((yhat-ytest)**2)))
```

```python
# Classification tree example with digits dataset
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets

# Load the diabetes dataset
random.seed(123456)
X, y = datasets.load_digits(return_X_y=True)
pos = random.sample(range(X.shape[0]), X.shape[0])

xtrain = X[pos[:1000],:]
ytrain = y[pos[:1000]] 
xtest = X[pos[1000:],:]
ytest = y[pos[1000:]] 

dtr = DecisionTreeClassifier(max_depth=10,min_samples_split=5)
dtr.fit(xtrain, ytrain)
yhat = dtr.predict(xtest)
count = 0
for i in range(ytest.shape[0]):
    if ytest[i]==yhat[i]:
        count=count+1
print(100*count/ytest.shape[0])

# Random Forest Regression example with diabetes dataset
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestRegressor(n_estimators=100,max_depth=10,min_samples_split=5)
rfc.fit(xtrain, ytrain)
yhat = rfc.predict(xtest)
for i in range(ytest.shape[0]):
    if ytest[i]==yhat[i]:
        count=count+1
print(100*count/ytest.shape[0])
```
